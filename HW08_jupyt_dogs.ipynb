```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn import tree
```

```python
# Look at the train data

train = pd.read_csv("./dogs_n_cats.csv")  
print(train.shape)
train.head()
```

```python
# checking if dataset has some null values
train.isnull().sum()
```

```python
# Look at the test data

import json

with open('dataset_209691_15.json', 'r') as json_file:  
    data = json.load(json_file)
    
test = pd.DataFrame.from_dict(data)
test.head()
```

```python
# We would like to predict if the item from test_data is "dog" based on train_data
# 1) Create a model/classificator

X = train.drop('Вид', axis=1)
y = train['Вид']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.20,
                                                    random_state=42)

X_train.shape
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test, y_test)
```

```python
# Estimating the model

y_pred = clf.predict(X_test)  
```

```python
from sklearn.metrics import classification_report, confusion_matrix 

print(confusion_matrix(y_test, y_pred))  
```

```python
print(classification_report(y_test, y_pred))
```

```python
(y_pred == y_test).sum()
```

```python
# Now lets predict our test_data based on model

X_test_data = test
y_pred_test = clf.predict(X_test_data)
```

```python
X_test_data['Вид'] = y_pred_test
```
